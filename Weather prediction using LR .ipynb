```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Summary of Weather.csv')
df
```

```python
X = df['MinTemp']
Y = df['MaxTemp']
```

```python
plt.scatter(X,Y)
```

```python
X = X.values.reshape(X.values.size,1)
```

```python
from sklearn.linear_model import LinearRegression
xr = LinearRegression()
xr.fit(X,Y)
```

```python
yp = xr.predict(X)
```

```python
xr.score(X,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X, yp,color = 'k')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 40,test_size=.2)
```

```python
sas = LinearRegression()
sas.fit(x_train ,y_train)
```

```python
y_pred = sas.predict(x_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
r2s = r2_score(y_test,y_pred)
r2s
```

```python
rmse = np.sqrt(mse)
rmse
```
